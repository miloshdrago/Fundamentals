```python
# apply the methods on the dataframe
imagedata_aggr = imagedata.groupby('user_id').mean()
```

```python
with pd.option_context('display.max_columns', None):
    display(imagedata_aggr.head())
```

Now we merge with the survey data

```python
survey_df = pd.read_pickle(r'visual_well_being_pickle_format/survey.pickle')
```

```python
survey_df.drop("index", 1, inplace= True)
```

```python
survey_df.columns
```

```python
imagedata_aggr.index = imagedata_aggr.index.astype("int64")
```

```python
# we can now merge the survey df with the feature df
total_df = pd.merge(survey_df, imagedata_aggr,  how="right",left_on ="insta_user_id",  right_on="user_id")
print(total_df.shape, imagedata_aggr.shape, survey_df.shape)

# save the merged pickle
total_df.to_pickle("dataset/total-DO-NOT-USE.pickle")
```

```python
# Split up into target and feature df
target_df = total_df[survey_df.columns]
feat_df = total_df[imagedata_aggr.columns]
```

```python
# Split of validation set
X_train, X_validate, y_train, y_validate = train_test_split(feat_df, target_df,
                                                               test_size = 0.10, random_state=42)
```

```python
X_train.shape, X_validate.shape, y_train.shape, y_validate.shape
```

```python
# Save to pickles
X_train.to_pickle("dataset/train_feat.pickle")
X_validate.to_pickle("dataset/validate_feat.pickle")
y_train.to_pickle("dataset/train_target.pickle")
y_validate.to_pickle("dataset/validate_target.pickle")
```
