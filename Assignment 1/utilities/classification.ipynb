```python
# dependencies
import pymongo
import pandas as pd
import nltk
from Sanitizer import Sanitizer
```

```python
client = pymongo.MongoClient()
db = client.fds_twitter
col = db.messages
itr = col.find(projection=[('id'), ('text'), ('entities.hashtags')])
```

```python
msgs = []
for found in itr:
    msgs.append({'id': found['id'], 'tweet': found['text'], 'hashtags': found['entities']['hashtags']})
```

# Noticable hastags
Grouped using a LDA model

* (0, '0.104*"Trump" + 0.033*"TrumpPence16"')
* (1, '0.049*"HillaryClinton" + 0.044*"NeverTrump"')

```python
df = pd.DataFrame(msgs)
```

```python
sanitizer = Sanitizer()
sanitized_df = df['tweet'].apply(sanitizer.sanitize)
```

```python
sanitized_df = pd.DataFrame(sanitized_df)
sanitized_df['id'] = df['id']
sanitized_df['hashtags'] = df['hashtags']
sanitized_df.set_index("id")
sanitized_df
```

```python
sanitized_df['id'] = sanitized_df['id'].apply(str)
sanitized_df
```

```python
#sanitized_df['hashtags'] = sanitized_df['hashtags'].apply(lambda x: [n['text'] for n in x])
sanitized_df
```

```python
trump_tags = ["Trump", "TrumpPence16", "NeverHillary"]
sanitized_df.loc[:, 'trump'] = [len(set(x).intersection(trump_tags)) > 1 for x in sanitized_df['hashtags']]
sanitized_df
```

```python
hillary_tags = ["HillaryClinton", "NeverTrump", "ImWithHer"]
sanitized_df.loc[:, "hillary"] = [len(set(x).intersection(hillary_tags)) > 1 for x in sanitized_df['hashtags']]
sanitized_df
```

```python
trump_tweets = sanitized_df.loc[sanitized_df['trump'] == True]
trump_tweets = trump_tweets['tweet'].tolist()
len(trump_tweets)
```

```python
hillary_tweets = sanitized_df.loc[sanitized_df['hillary'] == True]
hillary_tweets = hillary_tweets['tweet'].tolist()
len(hillary_tweets)
```

```python
trump_training_set = [(t, 'trump') for t in trump_tweets[:1000] if len(t) > 0]
hillary_training_set = [(t, 'hillary') for t in hillary_tweets[:1000] if len(t) > 0]
all_tweets = trump_training_set + hillary_training_set
len(trump_training_set), len(hillary_training_set)
```

```python
words = []
for (tweet, person) in all_tweets:
    words.extend(tweet)
word_features = nltk.FreqDist(words).keys()
len(word_features)
```

```python
def extract_features(message):
    words = set(message)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = word in words
    return features
```

```python
training_set = nltk.classify.apply_features(extract_features, all_tweets)
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
classifier.classify(extract_features(sanitizer.sanitize("i do not agree to what hillary is doing, i vote for trump")))
```

```python
classifier.classify(extract_features(sanitizer.sanitize("I am voting for hillary")))
```

```python
trump_correct = 0
trump_hillary = 0
hillary_correct = 0
hillary_trump = 0

for (idx, row) in sanitized_df.iterrows():
    if idx > 1000:
        break
    classified = classifier.classify(extract_features(row['tweet']))
    is_hillary = row['hillary']
    is_trump = row['trump']
    if classified == "trump" and is_trump:
        trump_correct += 1
    elif classified == "hillary" and is_trump:
        trump_hillary += 1
    elif classified == "hillary" and is_hillary:
        hillary_correct += 1
    elif classified == "trump" and is_hillary:
        hillary_trump += 1
```
