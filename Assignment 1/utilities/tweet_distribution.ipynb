```python
# dependencies
import os
import json
import pandas as pd
from pymongo import MongoClient
from collections import Counter
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, MultiPolygon
from shapely.affinity import scale
from descartes import PolygonPatch
from ast import literal_eval
%matplotlib inline
```

```python
# mongo connenction setup
client = MongoClient()
database = client.fds_twitter
collection = database.messages
collection.create_index([('id', 1)])
```

```python
places = collection.find({ "place.place_type": "city", "place.country_code": "US" }, projection=['id', 'place.bounding_box.coordinates'])
```

```python
# a container for all polygons
polygons = []

# iterate over all places
for place in places:
    polygons.append(PolygonPatch(Polygon(place['place']['bounding_box']['coordinates'][0])))
```

```python
# we need a figure and axis for the plot
fig = plt.figure(figsize=(30, 30))
ax = fig.gca()

# add all polygons to the figure
for poly in polygons:
    ax.add_patch(poly)

# we need to scale the axis so all
# places are in view
ax.axis('scaled')

# show without axes
plt.axis('off')
plt.show()
```

```python
coordinates = []
for place in places:
    coordinates.append(place['place']['bounding_box']['coordinates'][0])
```

```python
from collections import Counter
stringifed = [str(l) for l in coordinates]
```

```python
counted_coords = Counter(stringifed)
len(counted_coords)
```

```python
shapefile = '../../datasets/states_21basic/states.geojson'
os.path.isfile(shapefile)
```

```python
with open(shapefile, 'r') as rf:    
    data = json.load(rf)
    
BLUE = "steelblue"
fig = plt.figure(figsize=(50, 50))
ax = fig.gca()

for feature in data['features']:
    geometry = feature['geometry']
    if geometry['type'] == 'Polygon':
        poly = geometry
        ppatch = PolygonPatch(poly, fc=BLUE, ec=BLUE,  alpha=0.5, zorder=2)
        ax.add_patch(ppatch)
    elif geometry['type'] == 'MultiPolygon':
        for polygon in geometry['coordinates'][0]:
            poly = Polygon(polygon)
            ppatch = PolygonPatch(poly, fc=BLUE, ec=BLUE, alpha=0.5, zorder=2)
            ax.add_patch(ppatch)
    else:
        print('Don\'t know how to draw :', geometry['type'])

for coord, weight in counted_coords.items():
    polygon = Polygon(literal_eval(coord))
    scaled = scale(polygon, weight*0.001, weight*0.001)
    patch = PolygonPatch(scaled, fc="tomato", alpha=0.5)
    ax.add_patch(patch)

# we need to scale the axis so all
# places are in view
ax.axis('scaled')

# show without axes
plt.axis('off')
plt.show()
```
