## Latent Dirichlet Allocation
The next section will attempt on topic analysis using the builtin LDA model from the `gensim` library and a Mallet version of that model.

Here is a short description of how the LDA model tries to find topics:

1. Assume there are *k* topics across all of the documents
2. Distribute these *k* topics across document *m* (this distribution is known as **Î±** and can be symmetric or asymmetric) by assigning each word a topic.
3. For each word *w* in document *m*, assume its topic is wrong but every other word is assigned the correct topic.
4. Probabilistically assign word *w* a topic based on two things:
    - what topics are in document *m*
    - how many times word *w* has been assigned a particular topic across all of the documents (this distribution is called **Î²**.
5. Repeat this process a number of times for each document and youâ€™re done!

```python
# we need a Lda model to do the analysis with, for that we can
# use the Lda model provided by the gensim library
Lda = gensim.models.ldamodel.LdaModel

# we also need the Mallet Lda model. this model implementend a different algorithm, but
# often returns better results than the builtin gensim model. this is proven later on
mallet_path = './mallet-2.0.8/bin/mallet'
LdaMallet = gensim.models.wrappers.LdaMallet

# we are only interested in the sanitized tweets, so we need to convert that series
# to a list of a documents
documents = subdf['sanitized_text'].tolist()

# we need a dictionary of words that are mapped to ids, so that the model can track
# each word
ids2docs = gensim.corpora.Dictionary(documents)

# the lda model uses a document term frequency matrix to calculate the probability 
# of a word belonging to a topic based on how often it occurs in other documents, 
# given a probability that it belongs to a certain topic
corpus = [ids2docs.doc2bow(doc) for doc in documents]

len(documents), len(ids2docs.keys()), len(corpus)
```

```python
async def train_model(Model, *args, **kwargs):
    """
    Function to wrap around a timer around a model.
    @param  mixed     Model to train.
    @param  variadic  Arguments for the model.
    @param  variadic  Keyworded arguments for the model.
    @return mixed
    """
    return Model(*args, **kwargs)
```

# First model
Now that we have a document term frequency matrix and a model, we can train a model.\
We (want to) assume/expect two topics: hillary and trump.

However, often, political statements contain many complex words to describe a variety of topics, so
it is difficult to predict what number of topics actually exist. Therefore, multiple parameters
are used to compare them to find out what parameters are best.

A method to calculate how good a model is, is through calculating the coherence. This is what is
used to determine what model scored best.

```python
async def model_coherence(model, docs, dct):
    """
    Function to calculate the coherence of a model based on a list
    of documents and a dictionary of terms.
    @param  mixed                      Model to test.
    @param  list                       List of documents.
    @param  gensim.corpora.Dictionary  Dictionary of words mapping terms to ids.
    @return int
    """
    coherence_model_lda = gensim.models.CoherenceModel(model=model, texts=docs, dictionary=dct, coherence='c_v')
    return coherence_model_lda.get_coherence()
```

# Test models
Two functions are defined to help test the models: `model_coherence` and `test_model_coherence`.
In combination with a list of *k* topics, different models can be tested.

```python
# we need to define a set of steps that each represents the number of topics
ktopics = [2, 4, 6, 8, 10, 15, 20]
```

```python
async def test_models():
    """
    Function to test LDA models on a list of *k* number of topics.
    @return  dict,dict
    """
    # we need a list of models
    models = {'builtin': [], 'mallet': []}

    # we need a list of coherence scores
    scores = {'builtin': [], 'mallet': []}
        
    async def task(k):

        # we need to wait until the models are trainined
        builtin, mallet = await asyncio.gather(
            train_model(Lda, corpus, id2word=ids2docs, num_topics=k, alpha='asymmetric', eta="auto", chunksize=100),
            train_model(LdaMallet, mallet_path, corpus=corpus, id2word=ids2docs, num_topics=k)
        )
        
        # we need to wait for the scores of the models
        score_builtin, score_mallet = await asyncio.gather(
            model_coherence(builtin, documents, ids2docs),
            model_coherence(mallet, documents, ids2docs)
        )
        
        # add the models
        models['builtin'].append(builtin)
        models['mallet'].append(mallet)

        # add the coherence score to the list of scores
        scores['builtin'].append(score_builtin)
        scores['mallet'].append(score_mallet)
    
    # wait for all the tasks to finish
    await asyncio.gather(*[task(k) for k in ktopics])
    
    # expose the list of models and scores
    return models, scores

# get the models and scores
start = time.time()
models, scores = await test_models()
print('testing models took: {}s'.format(time.time()-start))
models, scores
```

```python
# plot to show the progression of scores for the builtin model
plt.plot(ktopics, scores['builtin'], c='tomato', label='Builtin LDA')
plt.plot(ktopics, scores['mallet'], c='steelblue', label='Mallet LDA')
plt.title('Coherence score per number of topics')
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(title='Coherence score', loc='best')
plt.show()
```

### Test results
The graph above shows that the Mallet LDA model outputs better topic coherence scores than the gensim builtin model.\
Both show, however, that the around 20 topics the highest coherence is found.

```python
# we use a custom timer as a decorator gives false 
# results for asynchronous function executions
start = time.time()

# the best model was the one at k=20
best_model = await train_model(LdaMallet, mallet_path, corpus=corpus, id2word=ids2docs, num_topics=20)
print('testing best model took: {}s'.format(time.time()-start))
```

```python
# output the topics that the model found
for (idx, words) in best_model.print_topics(num_topics=20, num_words=4):
    print("\nTopic: {}\n".format(idx))
    print("Words: {}\n".format(words))
```

## Plotting
Using the best model, we can plot the topics in a graph to get a illustrative view of the model.

Each bubble on the left-hand side plot represents a topic. The larger the bubble, the more prevalent is that topic.

A good topic model will have fairly big, non-overlapping bubbles scattered throughout the chart instead of being clustered in one quadrant.

A model with too many topics, will typically have many overlaps, small sized bubbles clustered in one region of the chart.

Alright, if you move the cursor over one of the bubbles, the words and bars on the right-hand side will update. These words are the salient keywords that form the selected topic.

```python
# we need to wrap the mallet to be able to plot it
best_model_wrapped = gensim.models.wrappers.ldamallet.malletmodel2ldamodel(best_model)

# plot the model
vis = pyLDAvis.gensim.prepare(best_model_wrapped, corpus, ids2docs, mds='mmds')
vis
```

## Time seasonality
It may be interesting to see how the topics differ per time period,
to be able to do that, we need to split the tweets up in time

```python
# we need to setup a mongo client first
client = pymongo.MongoClient()
db = client.fds_twitter.messages

# we need the first tweet and last tweet to see the datetime period
first_tweet = db.find(projection=["timestamp_ms"]).limit(1)
last_tweet = db.find(projection=["timestamp_ms"]).sort("timestamp_ms", -1).limit(1)

# we need the first and last date
first_date = datetime.fromtimestamp(int(first_tweet[0]['timestamp_ms'])/1000)
last_date = datetime.fromtimestamp(int(last_tweet[0]['timestamp_ms'])/1000)
```

```python
first_date, last_date
```

```python
# we need two weeks in seconds
one_week = 60 * 60 * 24 * 7
two_weeks = one_week * 2

# get a date that comes one week(s) after the first date
first_weeks = datetime.fromtimestamp(first_date.timestamp() + one_week)

# get a date that comes one week(s) before the last date
last_weeks = datetime.fromtimestamp(last_date.timestamp() - one_week)
```

```python
first_weeks, last_weeks
```

```python
# now we can get all tweets for the first two weeks
first_weeks_tweets = db.find(filter={"timestamp_ms": {"$lt": str(first_weeks.timestamp()*1000) }}, projection=["id", "text"])
last_weeks_tweets = db.find(filter={"timestamp_ms": {"$gt": str(last_weeks.timestamp()*1000) }}, projection=["id", "text"])
```

```python
# convert everything to lists
first_weeks_tweets = list(first_weeks_tweets)
last_weeks_tweets = list(last_weeks_tweets)
len(first_weeks_tweets), len(last_weeks_tweets)
```

```python
# we need to convert it to dataframes so we can easily analyse is
df_first = pd.DataFrame(first_weeks_tweets).iloc[:2000]
df_last = pd.DataFrame(last_weeks_tweets).iloc[:2000]
df_first.head(), df_last.head()
```

```python
# we need a sanitizer
sanitizer = Sanitizer()

# copy before modifying
df_first_sanitized = df_first.copy()
df_last_sanitized = df_last.copy()

# we need to sanitize all messages
df_first_sanitized.loc[:, 'text'] = df_first_sanitized.loc[:, 'text'].apply(lambda x: [t for t in sanitizer.sanitize(x) if len(t) >= 3])
df_last_sanitized.loc[:, 'text'] = df_last_sanitized.loc[:, 'text'].apply(lambda x: [t for t in sanitizer.sanitize(x) if len(t) >= 3])

# drop all empty columns
df_first_sanitized_clean = df_first_sanitized[df_first_sanitized['text'].map(lambda x: len(x) > 0)]
df_last_sanitized_clean = df_last_sanitized[df_last_sanitized['text'].map(lambda x: len(x) > 0)]

# output as test
df_first_sanitized_clean.head(), df_last_sanitized_clean.head()
```

```python
def model_corpus(series):
    """
    Function that converts a series into collections we can use
    for the model.
    @param  list  Series containing documents / tweets
    @return list,dict,list
    """
    
    # we are only interested in the sanitized tweets, so we need to convert that series
    # to a list of a documents
    documents = series

    # we need a dictionary of words that are mapped to ids, so that the model can track
    # each word
    ids2docs = gensim.corpora.Dictionary(documents)

    # the lda model uses a document term frequency matrix to calculate the probability 
    # of a word belonging to a topic based on how often it occurs in other documents, 
    # given a probability that it belongs to a certain topic
    corpus = [ids2docs.doc2bow(doc) for doc in documents]
    
    # expose the collections
    return documents, ids2docs, corpus
```

```python
# get the corpuses for the models for the week dataframes
first_corpus = model_corpus(df_first_sanitized_clean['text'].tolist())
last_corpus = model_corpus(df_last_sanitized_clean['text'].tolist())
```

```python
# train the models
first_model = await train_model(LdaMallet, mallet_path, corpus=first_corpus[2], id2word=first_corpus[1], num_topics=20)
last_model = await train_model(LdaMallet, mallet_path, corpus=last_corpus[2], id2word=last_corpus[1], num_topics=20)
```

```python
# output the coherence to check if we have a reasonable model
print('first model coherence', await model_coherence(first_model, first_corpus[0], first_corpus[1]))
print('last model coherence', await model_coherence(last_model, last_corpus[0], last_corpus[1]))
```
