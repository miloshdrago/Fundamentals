```python
# dependencies
import pymongo
import pandas as pd
import json
```

```python
# setup
client = pymongo.MongoClient()
database = client.fds_twitter
collection = database.messages
collection.create_index([('user.id', 1)])
```

```python
authors = collection.find(projection=["user.id", "user.name"])
df = pd.DataFrame(authors)
df
```

```python
users = pd.DataFrame([pd.Series(row) for row in df['user']])
```

```python
users
```

```python
unique_users = users.drop_duplicates(subset=["id"])
unique_users
```

```python
unique_users_sorted = unique_users.sort_values('id')
unique_users_sorted
```

```python
unique_users_sorted.to_csv('../datasets/users.csv')
```
